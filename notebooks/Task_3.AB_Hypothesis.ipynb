##### 1.Risk Differences Across Provinces

##### Metrics Selection:
I used Total Claims to measure risk differences across provinces, as it directly reflects the financial impact of claims. Total Premium measures revenue, and Margin indicates profitability, but neither directly quantifies risk. Total Claims aligns with our goal of assessing risk variations.

##### Data Segmentation:
Provinces were categorized into:

High Risk: Above-average Total Premium.

Low Risk: Below-average Total Premium.

##### Statistical Testing:
A t-test was used to compare the means of Total Claims between high-risk and low-risk provinces, as it is suitable for comparing two groups.

```python
# Categorize provinces based on TotalPremium
data_util = InsuranceDataUtils(df_subset)  # Use your df_subset or original DataFrame
high_risk_provinces, low_risk_provinces = data_util.categorize_provinces('TotalPremium')
t_stat, p_val, interpretation = data_util.test_risk_differences('TotalPremium', high_risk_provinces, low_risk_provinces)
data_util.print_summary(t_stat, p_val, interpretation)
```
